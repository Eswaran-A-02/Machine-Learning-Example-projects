```python
x = df.drop('Class',axis=1)
y = df['Class']
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,train_size=0.8,random_state=42)
xtrain
```

```python
pipeline = Pipeline(
    steps=[
        ('model', SVC())
    ]
)
pipeline.fit(xtrain,ytrain)
```

```python
xtrain_pred1 = pipeline.predict(xtrain)
print(classification_report(ytrain,xtrain_pred1))
```
