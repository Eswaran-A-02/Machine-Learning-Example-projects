```python
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv(r"C:\Users\eswar\OneDrive\Desktop\ML_example\spam_s.csv",encoding_errors='ignore')
df
```

```python
# df.drop(columns=(['Unnamed:2'],['Unnamed:3'],['Unnamed:4']),axis=1,inplace=True)
```

```python
x = df['v2']
y = df['v1']
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,train_size=0.8,random_state=42)
```

```python
pipelinee = Pipeline(
    steps =[
        ('vectors',CountVectorizer()),
        ('model',MultinomialNB())
    ]
)
pipelinee.fit(xtrain,ytrain)
```

```python
# pred = pipelinee
```

```python
xtrain_pred = pipelinee.predict(xtrain)
print(classification_report(ytrain,xtrain_pred))
```

```python
xtest_pred = pipelinee.predict(xtest)
print(classification_report(ytest,xtest_pred))
```

# tf&idf

```python
pipeline = Pipeline(
    steps =[
        ('vectors',TfidfVectorizer()),
        ('model',MultinomialNB())
    ]
)
pipeline.fit(xtrain,ytrain)
```

```python
xtrain_pred1 = pipeline.predict(xtrain)
print(classification_report(ytrain,xtrain_pred1))
```

```python
xtest_pred1 = pipeline.predict(xtest)
print(classification_report(ytest,xtest_pred1))
```

# logistic regression

```python
pipeline1 = Pipeline(
    steps =[
        ('vectors',TfidfVectorizer()),
        ('model',LogisticRegression())
    ]
)
pipeline1.fit(xtrain,ytrain)
```

```python
xtrain_pred2 = pipeline1.predict(xtrain)
print(classification_report(ytrain,xtrain_pred2))
```

```python

xtest_pred2 = pipeline1.predict(xtest)
print(classification_report(ytest,xtest_pred2))
```
